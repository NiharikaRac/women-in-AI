```python
#calcuating the no.of upper and lower case alphabets in given string

def calc_string(x):
 lower=0
 upper=0
 for i in x:
  if i.isupper()==True:
   upper += 1
  else:
   lower += 1
  return lower,upper
print(calc_string("she sells seashells by the seashore"))

```
